```python
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_excel("hdfc-bank-statement.xls",skiprows=20)) 
```

```python
# Drop the first row by index
df = df.iloc[1:].reset_index(drop=True)
df
```

```python
df = df.iloc[:449].reset_index(drop=True)
df
```

```python
df = df.dropna(subset=['Withdrawal Amt.', 'Deposit Amt.'], how='all')
```

```python
df = df.fillna(0)
df
```

```python
df['Withdrawal Amt.'] = pd.to_numeric(df['Withdrawal Amt.'], errors='coerce')
withdraw_sum = df['Withdrawal Amt.'].sum()
withdraw_sum
```

```python
df['Deposit Amt.'] = pd.to_numeric(df['Deposit Amt.'], errors='coerce')
dep_sum = df['Deposit Amt.'].sum()
dep_sum
```

```python
def categorize(row):
    if row['Withdrawal Amt.'] != 0:
        des = str(row['Narration']).lower()
        if "prepaid recharge" in des or "gpaybillpay-rchrg" in des:
            return "Recharge"
        elif "phonepewallettopup" in des:
            return "UPI Wallet Transfer"
        elif "moderncoeedu" in des or "modern college" in des or "unipune" in des:
            return "University Expenses"
        elif "zomato" in des:
            return "Food Order"
        elif "upi" in des:
            return "UPI Spending"
        elif "atm" in des:
            return "ATM Cash Withdrawal"
        elif "amb chrg" in des or "instaalertchg" in des or "debit card annual fee" in des:
            return "Bank Charges"
        else:
            return "Other"
```

```python
df['Category'] = df.apply(categorize, axis=1)
df
```

```python
df.to_excel("hdfc-bank-categorized.xlsx", index = False, engine = "openpyxl")
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%y')
df
```

```python
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df
```

```python
# Export the DataFrame to a CSV file
df.to_csv("bank_statement_cleaned.csv", index=False)
```
